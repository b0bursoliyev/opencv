```python
chess = cv2.imread('rasm/chess2.jpg')
gray = cv2.cvtColor(chess,cv2.COLOR_RGB2GRAY)
edg = cv2.Canny(chess,70,200)
plt.subplot(121)
plt.imshow(edg,cmap='gray')
plt.subplot(122)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
gray = np.float32(gray)
corner = cv2.cornerHarris(gray,5,5,0.04)
corner = cv2.dilate(corner,None)
plt.imshow(corner,cmap='gray')
```

```python
thresh = 0.01*corner.max()
for j in range(0,corner.shape[0]):
    for i in range(0,corner.shape[1]):
        if corner[j,i] > thresh:
            cv2.circle(chess,(i,j),3,(255,0,5),2)
plt.imshow(chess)
```

```python
dog = cv2.imread('rasm/game1.jpg')
dog = cv2.cvtColor(dog,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(dog,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
_,binary = cv2.threshold(gray,230,255,cv2.THRESH_BINARY_INV)
print(_)
plt.subplot(121)
plt.imshow(binary,cmap='gray')
```

```python
counter , soni = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
rasm1 = np.copy(dog)
rasm1 = cv2.drawContours(rasm1,counter,-1,(0,255,0),3)
plt.imshow(rasm1)
```

```python
a = cv2.imread('rasm/paint.png')
a = cv2.cvtColor(a,cv2.COLOR_BGR2GRAY)
plt.imshow(a,cmap='gray')
```

```python
kernel = np.ones((7,7),np.uint8)

dilate = cv2.dilate(a,kernel,iterations=1)
plt.subplot(121)
plt.imshow(dilate,cmap='gray')
plt.subplot(122)
plt.imshow(a,cmap='gray')
```

```python
erosion = cv2.erode(a,kernel,iterations=1)
plt.imshow(a,cmap='gray')
```

```python
plt.figure(figsize=(30,10))
plt.subplot(131)
plt.imshow(erosion,cmap='gray')
plt.subplot(132)
plt.imshow(a,cmap='gray')
plt.subplot(133)
plt.imshow(dilate,cmap='gray')
```

```python
yopiq = cv2.morphologyEx(a,cv2.MORPH_CLOSE,kernel)
plt.imshow(yopiq,cmap='gray')
```

```python
ochiq = cv2.morphologyEx(a,cv2.MORPH_OPEN,kernel)
plt.imshow(ochiq,cmap='gray')
```
