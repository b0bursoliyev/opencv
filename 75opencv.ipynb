```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image  as mimg
```

```python
car = cv2.imread("C:/Users/Lucas/Desktop/Suniy_intelkt/rasm/car_green_screen.jpg")
car = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
plt.imshow(car)
plt.show()
```

```python
car_copy = np.copy(car)
car_copy.shape
```

```python
rang1 = np.array([0,250,0])
rang2 = np.array([250,255,250])
masca = cv2.inRange(car_copy,rang1,rang2)
masca = cv2.resize(masca, (575,450))
car_copy = cv2.resize(car_copy, (575,450))
plt.imshow(masca,cmap='gray')
plt.show()
```

```python
%matplotlib inline
car_copy[masca!=0] = [0,0,0]
plt.imshow(car_copy)
plt.show()
```

```python
car = cv2.imread("C:/Users/Lucas/Desktop/Suniy_intelkt/rasm/car_green_screen.jpg")
car_copy = cv2.cvtColor(car,cv2.COLOR_BGR2RGB)
rang1 = np.array([0,250,0])
rang2 = np.array([250,255,250])
masca = cv2.inRange(car_copy,rang1,rang2)
masca = cv2.resize(masca, (575,450))
car_copy = cv2.resize(car_copy, (575,450))
car_copy[masca!=0] = [0,0,0]
sky = cv2.imread("C:/Users/Lucas/Desktop/Suniy_intelkt/rasm/sky.jpg")
sky = cv2.cvtColor(sky,cv2.COLOR_BGR2RGB)
sky = cv2.resize(sky, (575,450))
sky[masca==0] = [0,0,0]
tayyor1 = sky + car_copy
plt.imshow(tayyor1)
plt.show()
```
