```python
x = np.arange(0,2*np.pi,0.01)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
a = np.arange(0,4*np.e,0.01)
b = np.exp(a)
plt.plot(a,b)
plt.show()
```

```python
%matplotlib inline
x = np.arange(0,10)
y = 2*x+1
plt.plot(x,y)
plt.show()
```

```python
%matplotlib inline
x = np.arange(0,10,1)
y = np.arange(11,21,1)
plt.plot(x,y,'ro')
plt.title("ASDF")
plt.xlabel("X o'qi")
plt.ylabel("Y o'qi")
plt.axis([-5,15,-7,27])
plt.show()
```
