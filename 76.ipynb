```python
import os
import glob
import matplotlib.image as mimg
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
manzil = 'tun_kun/test/'
def starnartlash():
    rasm_list = []
    rasm_type = ['day','night']
    for img_type in rasm_type:
        for fayl in glob.glob(os.path.join(manzil,img_type,'*.jpg')):
            img = mimg.imread(fayl)
            img = cv2.resize(img, (720,1080))
            if img is not None:
                rasm_list.append((img,img_type))
    return rasm_list
rasmlar = starnartlash()
```

```python
manzil = 'rasm/'
def day(manzil):
    i=1
    rasm_day = []
    rasm_night = []
    for fayl in glob.glob(os.path.join(manzil,'*.jpg')):
        img = mimg.imread(fayl)
        plt.imshow(img)
        if img is not None:
            day = np.sum(img[img>=127])
            night = np.sum(img[img<127])
            day1 = np.std(img)
            night1 = np.std(img)
            day2 = np.mean(img)
            night2 = np.mean(img)
            if day2>=127 and day1<50:
                path = 'rasm/kun/'
                cv2.imwrite(os.path.join(path, f'{i}-rasm.jpg'), img)
                rasm_day.append(img)
                i+=1
            elif day2<127 and day1>50:
                path = 'rasm/tun/'
                cv2.imwrite(os.path.join(path, f'{i}-rasm.jpg'), img)
                rasm_night.append(img)
                i+=1
    return rasm_day
rasmlar = day(manzil)
```

```python
#Encode  day - 0,night - 1
def encode():
    for i in range(len(rasmlar)):
        if rasmlar[i][1]=='day':print(0)
        elif rasmlar[i][1]=='night':print(1)
encode()
```
